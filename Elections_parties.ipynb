### УИКи №5*** - ДЭГ в Москве

```python
df.sort_values('reg_voters', ascending=False).head()
```

```python
vrn_oblst = df.loc[df.reg == 'Воронежская область'].copy()
vrn_oblst
```

```python
vrn_oblst['%voted'] = (vrn_oblst.mobile_box_ballots + vrn_oblst.station_box_ballots) / vrn_oblst.reg_voters
```

```python
parties = list(df.columns)[-15:-1]
for party in parties:
    foo = '%' + str(party)
    vrn_oblst[foo] = vrn_oblst[party] / (vrn_oblst.mobile_box_ballots + vrn_oblst.station_box_ballots)
```

```python
vrn_oblst['%outsiders'] = vrn_oblst.outside_ballots / (vrn_oblst.outside_ballots + vrn_oblst.inside_ballots + vrn_oblst.preterm_ballots)
```

```python
vrn_oblst['%destroyers'] = vrn_oblst.invalid_ballots / (vrn_oblst.outside_ballots + vrn_oblst.inside_ballots + vrn_oblst.preterm_ballots)
```

```python
vrn_oblst.sort_values('%kprf',ascending=False).loc[:,['tik','uik','reg_voters','%voted','%outsiders','%destroyers','%er','%kprf']].head(50)
```

```python
vrn_oblst['total_ballots'] = (vrn_oblst.outside_ballots + vrn_oblst.inside_ballots + vrn_oblst.preterm_ballots)
vrn_oblst['total_ballots'].mean()
```

```python
vrn_oblst['dot_size'] = vrn_oblst.total_ballots / vrn_oblst['total_ballots'].mean()
```
